```python
import random

class Node:
    def __init__(self):
        self.parent = None
        self.children = []
```

```python

class Tree:
    def __init__(self):
        self.root = Node()
        self.nodes = [self.root]  # Liste pour garder les références aux nœuds

    def add_node(self):
        # Choisir un nœud aléatoire parmi les nœuds existants
        random_index = random.randint(0, len(self.nodes) - 1)
        chosen_node = self.nodes[random_index]

        # Créer un nouveau nœud et le relier au nœud choisi
        new_node = Node()
        new_node.parent = chosen_node
        chosen_node.children.append(new_node)

        # Ajouter le nouveau nœud à la liste des nœuds
        self.nodes.append(new_node)

    def print_tree(self):
        def _print_node(node, prefix=""):
            # Afficher le nœud actuel
            print(prefix + str(id(node)))
            # Afficher les enfants du nœud actuel
            for i, child in enumerate(node.children):
                child_prefix = prefix + "  └─ " if i == len(node.children) - 1 else prefix + "  ├─ "
                _print_node(child, child_prefix)
        _print_node(self.root)
```

```python

# Fonction pour générer un arbre de taille n
def generate_tree(n):
    tree = Tree()
    for _ in range(1, n):  # Déjà un nœud, donc on ajoute n-1 nœuds
        tree.add_node()
    return tree

# Exemple d'utilisation
if __name__ == "__main__":
    tree_size = [10, 30, 50, 100, 1000, 10000]
    for size in tree_size:
        tree = generate_tree(size)
        print(f"Arbre de taille {size} généré avec succès.")
        print("Affichage de l'arbre:")
        tree.print_tree()
```
